```python
# patch "Upscale" error 
# to address 'Upsample' object has no attribute 'recompute_scale_factor' error
# patch upsampling.py (in colab is /usr/local/lib/python3.8/dist-packages/torch/nn/modules/upsampling.py)
# according to this https://github.com/ultralytics/yolov5/issues/6948#issuecomment-1075528897
import os
torch_path = os.popen("pip show torch").read()
torch_path = torch_path.split("Location: ")[1].split("\n")[0]
upsampling_py_path = torch_path+"/torch/nn/modules/upsampling.py"
file_read = open(upsampling_py_path, "r")
file_str = file_read.read()
replace_file_str = file_str.replace("recompute_scale_factor=self.recompute_scale_factor", "#recompute_scale_factor=self.recompute_scale_factor\n")
file_write = open(upsampling_py_path, "w")
file_write.write(replace_file_str)
print(upsampling_py_path+" File Patched")
```
